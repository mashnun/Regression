```python
import numpy as np
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as pt
```

```python
np.random.seed(42)
n_samples = 100
X = np.linspace(0, 10, 100)
rng = np.random.randn(n_samples) * 100
Y = X**3 + rng + 100 #here the rng is the error term
#Y = X**3 + 100
pt.figure(figsize=(12,10))
pt.scatter(X, Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression 
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X.reshape(-1, 1))
```

```python
lr = LinearRegression()
lr.fit(X.reshape(-1, 1), Y)
model_pred = lr.predict(X.reshape(-1, 1))

pt.figure(figsize=(12,10))
pt.scatter(X,Y);
pt.plot(X, model_pred)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, Y.reshape(-1, 1))
Y_pred = lin_reg_2.predict (X_poly)
```

```python
pt.figure(figsize=(12,10))
pt.scatter(X, Y)
pt.plot(X, Y_pred)
```

```python
import numpy as np
import pandas as pd
import sys
from sklearn.datasets import load_boston
import seaborn as sns
```

```python
boston_data = load_boston()
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
df.shape
df.head()
```

```python
Y = boston_data.target
X = df
```

```python
X
Y
```

```python
from sklearn.tree import DecisionTreeRegressor
X = df[['LSTAT']].values
```

```python
X
```

```python
tree = DecisionTreeRegressor(max_depth=5)
tree.fit(X, Y)
```

```python
sort_idx = X.flatten().argsort()
```

```python
pt.figure(figsize=(12,10))
pt.scatter(X[sort_idx], Y[sort_idx])
pt.plot(X[sort_idx], tree.predict(X[sort_idx]), color = 'k')
pt.xlabel('LSTAT')
pt.ylabel('MEDV')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.values
#Y = df['MEDV'].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_estimators=500, criterion='mse', random_state=42, n_jobs=-1)
```

```python
forest.fit(X_train, Y_train)
```

```python
Y_train_pred = forest.predict(X_train)
Y_test_pred = forest.predict(X_test)
```

```python
print("MSE train: {0:.4f}, test: {1:.4f}".format(mean_squared_error(Y_train, Y_train_pred), mean_squared_error(Y_test, Y_test_pred)))
```

```python
print("R^2 train: {0:.4f}, test: {1:.4f}".format(r2_score(Y_train, Y_train_pred), r2_score(Y_test, Y_test_pred)))
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
ada = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4), n_estimators=500, random_state=42)
```

```python
ada.fit(X_train, Y_train)
Y_train_pred = ada.predict(X_train)
Y_test_pred = ada.predict(X_test)
```

```python
print("MSE train: {0:.4f}, test: {1:.4f}".format(mean_squared_error(Y_train, Y_train_pred), mean_squared_error(Y_test, Y_test_pred)))
```

```python
print("R^2 train: {0:.4f}, test: {1:.4f}".format(r2_score(Y_train, Y_train_pred), r2_score(Y_test, Y_test_pred)))
```
